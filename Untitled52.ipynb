```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import datetime as dt

# ===== Load =====
rbsp = sio.loadmat("data/RBSP.mat")
rbsp_energy = sio.loadmat("data/RBSPenergy.mat")

T = rbsp["T"].flatten()
CountRate = rbsp["CountRate"]

E    = rbsp_energy["E"].flatten()
Emin = rbsp_energy["Emin"].flatten()
Emax = rbsp_energy["Emax"].flatten()
G    = rbsp_energy["G"].flatten()
dE = Emax - Emin

# Fix orientation if needed (N,10)
if CountRate.shape[0] == 10 and CountRate.shape[1] != 10:
    CountRate = CountRate.T

def matlab_datenum_to_datetime(datenum_array):
    return np.array([
        dt.datetime.fromordinal(int(d)) +
        dt.timedelta(days=float(d) % 1) -
        dt.timedelta(days=366)
        for d in datenum_array
    ])

time = matlab_datenum_to_datetime(T)

# ===== 1 day before storm: 2013-03-16 =====
start = dt.datetime(2013, 3, 16, 0, 0, 0)
end   = dt.datetime(2013, 3, 17, 0, 0, 0)
mask_day = (time >= start) & (time < end)

# ===== Differential flux (based on statement: G in sr^-1 cm^-2) =====
flux = CountRate * (G / dE)          # cm^-2 sr^-1 s^-1 keV^-1
flux = np.where(flux > 0, flux, np.nan)

flux_day = flux[mask_day, :]

# Mean spectrum (or median)
mean_flux = np.nanmean(flux_day, axis=0)
# mean_flux = np.nanmedian(flux_day, axis=0)

# (optional) check values
print("E (keV):", E)
print("mean_flux:", mean_flux)

# ===== Power-law fitting in log-log =====
def fit_powerlaw_loglog(Ev, Jv):
    m = np.isfinite(Ev) & np.isfinite(Jv) & (Ev > 0) & (Jv > 0)
    Ev = Ev[m]; Jv = Jv[m]
    idx = np.argsort(Ev)
    Ev = Ev[idx]; Jv = Jv[idx]

    x = np.log(Ev)
    y = np.log(Jv)

    a, b = np.polyfit(x, y, 1)   # y = a x + b
    A = np.exp(b)
    gamma = -a
    yhat = a*x + b
    sse_log = np.sum((y - yhat)**2)
    return A, gamma, sse_log, Ev, Jv

# Two populations: low-E (channels 1..6) and high-E (channels 7..10)
E_low,  J_low  = E[:6],  mean_flux[:6]
E_high, J_high = E[6:],  mean_flux[6:]

A1, g1, sse1, E_low_s,  J_low_s  = fit_powerlaw_loglog(E_low,  J_low)
A2, g2, sse2, E_high_s, J_high_s = fit_powerlaw_loglog(E_high, J_high)

print(f"Low-E population (ch1–6):  A1={A1:.3e}, gamma1={g1:.3f}, SSE_log={sse1:.3e}")
print(f"High-E population (ch7–10): A2={A2:.3e}, gamma2={g2:.3f}, SSE_log={sse2:.3e}")

# ===== Plot spectrum + fits =====
plt.figure()
plt.loglog(E, mean_flux, "o", label="Spectrum (mean, 2013-03-16)")

Egrid1 = np.logspace(np.log10(E_low_s.min()),  np.log10(E_low_s.max()), 200)
Egrid2 = np.logspace(np.log10(E_high_s.min()), np.log10(E_high_s.max()), 200)

plt.loglog(Egrid1, A1 * Egrid1**(-g1), "-", label=f"Fit low-E: γ1={g1:.2f}")
plt.loglog(Egrid2, A2 * Egrid2**(-g2), "-", label=f"Fit high-E: γ2={g2:.2f}")

plt.xlabel("Energy E [keV]")
plt.ylabel("Mean differential flux ⟨j(E)⟩ [cm$^{-2}$ sr$^{-1}$ s$^{-1}$ keV$^{-1}$]")
plt.title("Power-law fits (two energy populations)")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit

E = np.array([  37.3,   56.7,   80.4,  110.9,  184.9,  221.1,  846. , 1253. , 1535. , 1942. ])
J = np.array([3.77363303, 0.99793173, 0.26723171, 0.10216183, 0.03141635,
              0.01452189, 0.65875457, 0.46409967, 0.11199435, 0.0196571 ])

def powerlaw(E, A, gamma):
    return A * (E ** (-gamma))

def fit_loglog(Ev, Jv):
    x = np.log10(Ev)
    y = np.log10(Jv)
    b, a = np.polyfit(x, y, 1)   # y = a + b x
    A = 10**a
    gamma = -b
    return A, gamma

def fit_curvefit(Ev, Jv):
    # αρχικές τιμές από loglog (βοηθάει να συγκλίνει)
    A0, g0 = fit_loglog(Ev, Jv)
    popt, pcov = curve_fit(powerlaw, Ev, Jv, p0=[A0, g0], maxfev=20000)
    A, gamma = popt
    perr = np.sqrt(np.diag(pcov))  # 1-sigma uncertainties
    return A, gamma, perr

# splits
E_low,  J_low  = E[:6], J[:6]
E_high, J_high = E[6:], J[6:]

# --- Method 1: log-log linear ---
A1_ll, g1_ll = fit_loglog(E_low, J_low)
A2_ll, g2_ll = fit_loglog(E_high, J_high)

# --- Method 2: nonlinear LS in original space ---
A1_cf, g1_cf, err1 = fit_curvefit(E_low, J_low)
A2_cf, g2_cf, err2 = fit_curvefit(E_high, J_high)

print("LOG-LOG polyfit:")
print(f" Low-E :  A={A1_ll:.3e}, gamma={g1_ll:.3f}")
print(f" High-E:  A={A2_ll:.3e}, gamma={g2_ll:.3f}\n")

print("Nonlinear curve_fit (original space):")
print(f" Low-E :  A={A1_cf:.3e} ± {err1[0]:.3e}, gamma={g1_cf:.3f} ± {err1[1]:.3f}")
print(f" High-E:  A={A2_cf:.3e} ± {err2[0]:.3e}, gamma={g2_cf:.3f} ± {err2[1]:.3f}")
```

```python
import matplotlib.pyplot as plt

plt.figure()
plt.loglog(E, J, 'o', label="Spectrum (mean)")

Egrid1 = np.logspace(np.log10(E_low.min()), np.log10(E_low.max()), 200)
Egrid2 = np.logspace(np.log10(E_high.min()), np.log10(E_high.max()), 200)

plt.loglog(Egrid1, powerlaw(Egrid1, A1_ll, g1_ll), '-', label=f"loglog low: γ={g1_ll:.2f}")
plt.loglog(Egrid1, powerlaw(Egrid1, A1_cf, g1_cf), '--', label=f"curvefit low: γ={g1_cf:.2f}")

plt.loglog(Egrid2, powerlaw(Egrid2, A2_ll, g2_ll), '-', label=f"loglog high: γ={g2_ll:.2f}")
plt.loglog(Egrid2, powerlaw(Egrid2, A2_cf, g2_cf), '--', label=f"curvefit high: γ={g2_cf:.2f}")

plt.xlabel("Energy E [keV]")
plt.ylabel("⟨j(E)⟩ [cm$^{-2}$ sr$^{-1}$ s$^{-1}$ keV$^{-1}$]")
plt.title("Comparison of fitting methods (two populations)")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()
```

Μέθοδος 1 (log–log γραμμική παλινδρόμηση): ελαχιστοποιεί σφάλματα στον
log 𝑗
logj ⇒ δίνει έμφαση σε σχετικές αποκλίσεις (multiplicative noise). Είναι πολύ συνηθισμένη για power laws και σταθερή αριθμητικά.

Μέθοδος 2 (nonlinear LS στο j): ελαχιστοποιεί σφάλματα στο 𝑗
j ⇒ δίνει μεγαλύτερο βάρος στα μεγάλα j (χαμηλές ενέργειες), εκτός αν βάλεις weights. Πλεονέκτημα: μπορεί να σου δώσει και αβεβαιότητες από το covariance matrix.

Πρακτικά:

Τα γ των δύο μεθόδων συχνά είναι παρόμοια αν το power law είναι καλό μοντέλο.

Αν διαφέρουν, αυτό δείχνει ότι το fit επηρεάζεται από το “πού” μετράς το σφάλμα (log space vs linear space) ή/και ότι η περιοχή δεν είναι καθαρή power law (ιδίως στο high-E που έχεις μόνο 4 σημεία).

Εφαρμόστηκαν δύο μέθοδοι fitting: (i) γραμμική παλινδρόμηση σε log–log χώρο και (ii) μη γραμμικό least-squares fit στο αρχικό χώρο. Στις χαμηλές ενέργειες οι δύο μέθοδοι δίνουν παρόμοιες τιμές της φασματικής κλίσης (γ≈3.1–3.3), υποδεικνύοντας ότι το φάσμα περιγράφεται ικανοποιητικά από power law. Αντίθετα, στις υψηλές ενέργειες παρατηρείται σημαντική απόκλιση μεταξύ των δύο μεθόδων, γεγονός που αποδίδεται στο μικρό πλήθος δεδομένων και σε πιθανή καμπύλωση ή επιπλέον break στο MeV εύρος. Η σύγκριση αναδεικνύει τους περιορισμούς της απλής power-law προσέγγισης στις υψηλές ενέργειες

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import datetime as dt

# ---------- helpers ----------
def matlab_datenum_to_datetime(datenum_array):
    return np.array([
        dt.datetime.fromordinal(int(d)) +
        dt.timedelta(days=float(d) % 1) -
        dt.timedelta(days=366)
        for d in datenum_array
    ])

def fit_powerlaw_loglog(Ev, Jv):
    """Fit J = A * E^(-gamma) using log-log linear regression."""
    m = np.isfinite(Ev) & np.isfinite(Jv) & (Ev > 0) & (Jv > 0)
    Ev = Ev[m]; Jv = Jv[m]
    idx = np.argsort(Ev)
    Ev = Ev[idx]; Jv = Jv[idx]

    x = np.log10(Ev)
    y = np.log10(Jv)

    b, a = np.polyfit(x, y, 1)      # y = a + b x
    gamma = -b
    A = 10**a

    yhat = a + b*x
    sse_log = np.sum((y - yhat)**2)

    return A, gamma, sse_log

# ---------- load data ----------
rbsp = sio.loadmat("data/RBSP.mat")
ene  = sio.loadmat("data/RBSPenergy.mat")

T = rbsp["T"].flatten()
CountRate = rbsp["CountRate"]
time = matlab_datenum_to_datetime(T)

E    = ene["E"].flatten()
Emin = ene["Emin"].flatten()
Emax = ene["Emax"].flatten()
G    = ene["G"].flatten()
dE = Emax - Emin

# make CountRate (N,10)
if CountRate.shape[0] == 10 and CountRate.shape[1] != 10:
    CountRate = CountRate.T

# differential flux (based on statement: G in sr^-1 cm^-2)
flux = CountRate * (G / dE)  # cm^-2 sr^-1 s^-1 keV^-1
flux = np.where(flux > 0, flux, np.nan)

# ---------- function: spectrum for a period ----------
def spectrum_for_period(start, end, method="mean"):
    m = (time >= start) & (time < end)
    F = flux[m, :]
    if method == "median":
        return np.nanmedian(F, axis=0)
    return np.nanmean(F, axis=0)

# ---------- DEFINE YOUR 3 ONE-DAY PERIODS ----------
# ΒΑΛΕ ΕΔΩ τα start/end που επέλεξες (1 ημέρα το καθένα).
periods = [
    ("Main phase",     dt.datetime(2013, 3, 17, 0, 0, 0), dt.datetime(2013, 3, 18, 0, 0, 0)),
    ("Recovery phase", dt.datetime(2013, 3, 18, 0, 0, 0), dt.datetime(2013, 3, 19, 0, 0, 0)),
    ("Post-storm",      dt.datetime(2013, 3, 19, 0, 0, 0), dt.datetime(2013, 3, 20, 0, 0, 0)),
]

# ---------- run + plot ----------
results = []

plt.figure()
for name, start, end in periods:
    spec = spectrum_for_period(start, end, method="mean")   # ή "median"

    # two populations: ch1–6 and ch7–10
    A1, g1, sse1 = fit_powerlaw_loglog(E[:6],  spec[:6])
    A2, g2, sse2 = fit_powerlaw_loglog(E[6:],  spec[6:])

    results.append((name, start, end, A1, g1, sse1, A2, g2, sse2))

    # plot spectrum
    plt.loglog(E, spec, 'o-', label=f"{name}")

plt.xlabel("Energy E [keV]")
plt.ylabel("Mean differential flux ⟨j(E)⟩ [cm$^{-2}$ sr$^{-1}$ s$^{-1}$ keV$^{-1}$]")
plt.title("Energy spectra for 3 storm phases (one-day averages)")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()

# ---------- print parameters ----------
print("\n=== Power-law parameters (log-log fits) ===")
for (name, start, end, A1, g1, sse1, A2, g2, sse2) in results:
    print(f"\n{name}: {start}  to  {end}")
    print(f"  Low-E  (ch1–6, {E[0]:.1f}–{E[5]:.1f} keV):  A1={A1:.3e}, gamma1={g1:.3f}, SSElog={sse1:.3e}")
    print(f"  High-E (ch7–10,{E[6]:.1f}–{E[9]:.1f} keV):  A2={A2:.3e}, gamma2={g2:.3f}, SSElog={sse2:.3e}")

# ---------- optional: plot fits per period (one figure per period) ----------
for (name, start, end, A1, g1, sse1, A2, g2, sse2) in results:
    spec = spectrum_for_period(start, end, method="mean")

    plt.figure()
    plt.loglog(E, spec, "o", label="Spectrum")

    Egrid1 = np.logspace(np.log10(E[:6].min()), np.log10(E[:6].max()), 200)
    Egrid2 = np.logspace(np.log10(E[6:].min()), np.log10(E[6:].max()), 200)

    plt.loglog(Egrid1, A1 * Egrid1**(-g1), "-", label=f"Low-E fit: γ={g1:.2f}")
    plt.loglog(Egrid2, A2 * Egrid2**(-g2), "-", label=f"High-E fit: γ={g2:.2f}")

    plt.xlabel("Energy E [keV]")
    plt.ylabel("⟨j(E)⟩ [cm$^{-2}$ sr$^{-1}$ s$^{-1}$ keV$^{-1}$]")
    plt.title(f"{name} spectrum + power-law fits ({start.date()} to {end.date()})")
    plt.grid(True, which="both")
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# δεδομένα (βάλε τα δικά σου)
E = np.array([  37.3,   56.7,   80.4,  110.9,  184.9,  221.1,  846. , 1253. , 1535. , 1942. ])
J = np.array([3.77363303, 0.99793173, 0.26723171, 0.10216183, 0.03141635,
              0.01452189, 0.65875457, 0.46409967, 0.11199435, 0.0196571 ])

# splits
E_low,  J_low  = E[:6], J[:6]
E_high, J_high = E[6:], J[6:]

# power law params (από πριν)
A_low, gamma_low  = 2.135e5, 3.054
A_high, gamma_high = 1.638e12, 4.162

Jfit_low  = A_low  * (E_low  ** (-gamma_low))
Jfit_high = A_high * (E_high ** (-gamma_high))

def exp_model(E_keV, J0, E0):
    return J0 * np.exp(-E_keV / E0)

# exp fits
p0_low  = (np.max(J_low),  100.0)
p0_high = (np.max(J_high), 1000.0)

popt_low, pcov_low   = curve_fit(exp_model, E_low,  J_low,  p0=p0_low,  maxfev=20000)
popt_high, pcov_high = curve_fit(exp_model, E_high, J_high, p0=p0_high, maxfev=20000)

J0_low, E0_low   = popt_low
J0_high, E0_high = popt_high

print("Exponential fits:")
print(f"  Low-energy : J0 ≈ {J0_low:.3e},  E0 ≈ {E0_low:.2f} keV")
print(f"  High-energy: J0 ≈ {J0_high:.3e}, E0 ≈ {E0_high:.2f} keV")

# Plots
plt.figure()
plt.loglog(E_low, J_low, 'o', label="Low-energy data")
plt.loglog(E_low, Jfit_low, '-', label=f"Power law (γ ≈ {gamma_low:.2f})")
plt.loglog(E_low, exp_model(E_low, J0_low, E0_low), '--', label=f"Exp (E0 ≈ {E0_low:.1f} keV)")
plt.xlabel("Energy E [keV]")
plt.ylabel("⟨j(E)⟩")
plt.title("Low-energy: power law vs exponential")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()

plt.figure()
plt.loglog(E_high, J_high, 's', label="High-energy data")
plt.loglog(E_high, Jfit_high, '-', label=f"Power law (γ ≈ {gamma_high:.2f})")
plt.loglog(E_high, exp_model(E_high, J0_high, E0_high), '--', label=f"Exp (E0 ≈ {E0_high:.1f} keV)")
plt.xlabel("Energy E [keV]")
plt.ylabel("⟨j(E)⟩")
plt.title("High-energy: power law vs exponential")
plt.grid(True, which="both")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import scipy.io as sio
import datetime as dt

# =========================
# Helpers
# =========================
def matlab_datenum_to_datetime(datenum_array):
    datenum_array = np.asarray(datenum_array).astype(float)
    return np.array([
        dt.datetime.fromordinal(int(d)) +
        dt.timedelta(days=float(d) % 1) -
        dt.timedelta(days=366)
        for d in datenum_array
    ])

def fit_powerlaw_loglog(Ev, Jv):
    Ev = np.asarray(Ev, dtype=float)
    Jv = np.asarray(Jv, dtype=float)

    m = np.isfinite(Ev) & np.isfinite(Jv) & (Ev > 0) & (Jv > 0)
    Ev = Ev[m]; Jv = Jv[m]

    idx = np.argsort(Ev)
    Ev = Ev[idx]; Jv = Jv[idx]

    x = np.log10(Ev)
    y = np.log10(Jv)

    b, a = np.polyfit(x, y, 1)   # y = a + b x
    gamma = -b
    A = 10**a

    yhat = a + b*x
    sse_log = np.sum((y - yhat) ** 2)
    return A, gamma, sse_log

def spectrum_for_period(time, flux, start, end, method="mean"):
    m = (time >= start) & (time < end)
    F = flux[m, :]
    if F.size == 0:
        return None
    if method == "median":
        return np.nanmedian(F, axis=0)
    return np.nanmean(F, axis=0)

# =========================
# Load data
# =========================
rbsp = sio.loadmat("data/RBSP.mat")
ene  = sio.loadmat("data/RBSPenergy.mat")

T = rbsp["T"].flatten()
CountRate = rbsp["CountRate"]
time = matlab_datenum_to_datetime(T)

E    = ene["E"].flatten()
Emin = ene["Emin"].flatten()
Emax = ene["Emax"].flatten()
G    = ene["G"].flatten()
dE = Emax - Emin

# orientation (N,10)
if CountRate.shape[0] == 10 and CountRate.shape[1] != 10:
    CountRate = CountRate.T

# flux (based on statement: G in sr^-1 cm^-2)
flux = CountRate * (G / dE)
flux = np.where(flux > 0, flux, np.nan)

# =========================
# 4 phases (24h each)
# =========================
periods4 = [
    ("Pre-storm",      dt.datetime(2013, 3, 16, 0, 0, 0), dt.datetime(2013, 3, 17, 0, 0, 0)),
    ("Main phase",     dt.datetime(2013, 3, 17, 0, 0, 0), dt.datetime(2013, 3, 18, 0, 0, 0)),
    ("Recovery phase", dt.datetime(2013, 3, 18, 0, 0, 0), dt.datetime(2013, 3, 19, 0, 0, 0)),
    ("Post-storm",     dt.datetime(2013, 3, 19, 0, 0, 0), dt.datetime(2013, 3, 20, 0, 0, 0)),
]

# =========================
# Compute fits + poetic analysis
# =========================
phase_stats = []

for name, start, end in periods4:
    spec = spectrum_for_period(time, flux, start, end, method="mean")
    if spec is None:
        print(f"[WARN] No data for {name}")
        continue

    A1, g1, sse1 = fit_powerlaw_loglog(E[:6], spec[:6])
    A2, g2, sse2 = fit_powerlaw_loglog(E[6:], spec[6:])

    phase_stats.append({
        "name": name,
        "start": start,
        "end": end,
        "A1": A1, "g1": g1, "sse1": sse1,
        "A2": A2, "g2": g2, "sse2": sse2
    })

# helper for poetic text
def mood_from_gamma(g):
    if g >= 3.6:
        return "μαλακό και απότομο, σαν να σβήνει γρήγορα"
    if g >= 2.6:
        return "ισορροπημένο, με καθαρή κλίση"
    if g >= 1.8:
        return "πιο σκληρό, κρατάει ουρά"
    return "σκληρό και επίμονο, δεν αφήνει εύκολα την ενέργεια"

def compare(prev, cur, key):
    dg = cur[key] - prev[key]
    if abs(dg) < 0.15:
        return "σχεδόν δεν αλλάζει"
    if dg > 0:
        return "μαλακώνει (η κλίση μεγαλώνει)"
    return "σκληραίνει (η κλίση μικραίνει)"

print("\n==================== FIT PARAMETERS (4 phases) ====================")
for p in phase_stats:
    print(f"\n{p['name']}: {p['start']} to {p['end']}")
    print(f"  Low-E :  A1={p['A1']:.3e}, gamma1={p['g1']:.3f}, SSElog={p['sse1']:.3e}")
    print(f"  High-E:  A2={p['A2']:.3e}, gamma2={p['g2']:.3f}, SSElog={p['sse2']:.3e}")

print("\n==================== POETIC ANALYSIS ====================")

# per-phase poetic snapshot
for p in phase_stats:
    print(f"\n{p['name']} ({p['start'].date()}):")
    print(f"  Στις χαμηλές ενέργειες, το φάσμα είναι {mood_from_gamma(p['g1'])} (γ1={p['g1']:.2f}).")
    print(f"  Στις υψηλές ενέργειες, το φάσμα είναι {mood_from_gamma(p['g2'])} (γ2={p['g2']:.2f}).")

# transitions (how it changes phase-to-phase)
for i in range(1, len(phase_stats)):
    prev = phase_stats[i-1]
    cur  = phase_stats[i]
    print(f"\nΜετάβαση {prev['name']} → {cur['name']}:")
    print(f"  Low-E:  {compare(prev, cur, 'g1')} (γ1: {prev['g1']:.2f} → {cur['g1']:.2f})")
    print(f"  High-E: {compare(prev, cur, 'g2')} (γ2: {prev['g2']:.2f} → {cur['g2']:.2f})")

# small scientific closing line (still poetic-ish)
print("\nΚλείσιμο:")
print("  Όσο η καταιγίδα κορυφώνεται, το φάσμα αλλάζει ρυθμό σαν μουσική: άλλοτε σφίγγει, άλλοτε χαλαρώνει,")
print("  και στις υψηλές ενέργειες φαίνεται πιο εύθραυστο—σαν λεπτή γραμμή που την τραβούν μαζί επιτάχυνση και απώλειες.")
```
